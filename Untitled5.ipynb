```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url ='https://www.worldometers.info/coronavirus/'
html =requests.get(url).text
soup = BeautifulSoup(html)
```

```python
#get the required table
table =soup.find('table',attrs={'id':'main_table_countries_today'})

#get all the rows
rows =table.find_all('tr')

# store rows into list (data).
data =[]
for row in rows:
  data.append(row.text.strip().split('\n')[1:5])
```

```python
# convert list into DataFrame and Excluding empthy rows
df = pd.DataFrame(data[9:240], columns =data[0])
df.to_csv('covid19.csv')
```

```python
# checking the data
df.info()
```

```python
# Convert 'TotalCases' to numeric for sorting to visualize the top and least 10 countries according to total cases
df['TotalCases']=pd.to_numeric(df['TotalCases'].str.replace(',',''))
```

```python
#sort the DataFram by 'TotalCases' in descending order
df_sorted =df.sort_values(by='TotalCases', ascending=False)
```

```python
#get the top and the least 10 countries
top_10 =df_sorted.head(10)
least_10 =df_sorted.tail(10)
```

```python
 #Plotting top 10
top_10.plot(kind='bar',x='Country,Other',y='TotalCases')
```

```python
#Plotting least 10
least_10.plot(kind='bar',x='Country,Other',y='TotalCases')
```
